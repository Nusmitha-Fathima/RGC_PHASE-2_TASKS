### titanic.csv

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
missing_values = df.isna()
missing_values
```

```python
missing_count = df.isna().sum()
missing_count
```

```python
x=df["Age"].median()
df["Age"].fillna(x,inplace=True)
df
```

```python
x=df["Cabin"].mode()[0]
x
```

```python
df["Cabin"].fillna(x,inplace=True)
df
```

```python
x=df["Embarked"].mode()[0]
x
```

```python
 df["Embarked"].fillna(x,inplace=True)
```

```python
df
```

```python
missing_count = df.isna().sum()
missing_count
```
